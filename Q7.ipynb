```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from sklearn import svm
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.gca().set_aspect('equal', adjustable='box')
# 1是+ 2是-
x1=[2,2,3]
x1B=[2,2,3,2]
x1C=[2,2,3,2,-2]
x2=[-1,0,1,-1]
y1=[2,3,2]
y1B=[2,3,2,-2]
y1C=[2,3,2,-2,-2]
y2=[1,0,-1,0]
unknown=(0,2.5)
plt.scatter(x1,y1,label = "positive",c ="red",s = 200)
plt.scatter(x2,y2,label = "negative",c ="green",s = 200)
plt.scatter(unknown[0],unknown[1],label="unknown",c="blue",s=200)
x = np.linspace(-4.0, 4.0, 100)
y = np.linspace(-4.0, 4.0, 100)
X, Y = np.meshgrid(x,y)
F = -X - Y + 2
plt.contour(X,Y,F,[0])
plt.legend()
plt.title("Dataset A")
```
