```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from sklearn import svm
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.gca().set_aspect('equal', adjustable='box')
# 1是+ 2是-
x1=[2,2,3]
x1B=[2,2,3,2]
x1C=[2,2,3,2,-2]
x2=[-1,0,1,-1]
y1=[2,3,2]
y1B=[2,3,2,-2]
y1C=[2,3,2,-2,-2]
y2=[1,0,-1,0]
unknown=(0,2.5)
plt.scatter(x1,y1,label = "positive",c ="red",s = 200)
plt.scatter(x2,y2,label = "negative",c ="green",s = 200)
plt.scatter(unknown[0],unknown[1],label="unknown",c="blue",s=200)
x = np.linspace(-4.0, 4.0, 100)
y = np.linspace(-4.0, 4.0, 100)

X, Y = np.meshgrid(x,y)
F = -X - Y + 2
plt.contour(X,Y,F,[0])
plt.legend()
plt.title("Dataset A")
```

```python
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(x1B,y1B,label = "positive",c ="red",s = 200)
plt.scatter(x2,y2,label = "negative",c ="green",s = 200)
plt.scatter(unknown[0],unknown[1],label="unknown",c="blue",s=200)
plt.axvline(x=1.5, linestyle='-')
plt.title("Dataset B")
plt.legend()
```

```python
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(x1C,y1C,label = "positive",c ="red",s = 200)
plt.scatter(x2,y2,label = "negative",c ="green",s = 200)
plt.scatter(unknown[0],unknown[1],label="unknown",c="blue",s=200)
plt.title("Dataset C")
plt.legend()
plt.show()
```

```python
plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.gca().set_aspect('equal', adjustable='box')
x1CN=[i**2 for i in x1C]
y1CN=[i**2 for i in y1C]
x2N=[i**2 for i in x2]
y2N=[i**2 for i in y2]
plt.scatter(x1CN,y1CN,label = "positive",c ="red",s = 100)
plt.scatter(x2N,y2N,label = "negative",c ="green",s = 100)
plt.scatter(unknown[0],unknown[1],label="unknown",c="blue",s=100)
plt.title("Dataset C")
a = np.linspace(-15, 15, 100)

plt.plot(a, -a+5, '-r')

plt.legend()
plt.show()
# print("-的点坐标为")
# print(x2N)
# print(y2N)
# print("+的坐标为")
# print(x1CN)
# print(y1CN)
```

```python
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(x1C,y1C,label = "positive",c ="red",s = 200)
plt.scatter(x2,y2,label = "negative",c ="green",s = 200)
plt.scatter(unknown[0],unknown[1],label="unknown",c="blue",s=200)
plt.title("Dataset C")
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x,y)
F = X**2 + Y**2 - 5
plt.contour(X,Y,F,[0])
plt.legend()
plt.show()
```
